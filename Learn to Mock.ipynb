```python
# Much of this came from here: http://nbviewer.ipython.org/gist/jiffyclub/9235955 
import github3

def public(thing):
    """Print 'public' attributes of thing."""
    for x in dir(thing):
        if not x.startswith('_'):
            print(x)
            
# use your username and your preferred repository name:
my_username = 'csheldonhess'
my_repo_name = 'api-test'

# generate a token for yourself in GitHub; paste it in below    
g = github3.login(token='')

if g:
    print('Access granted.\n\n')
    
print('What can be done with g:')
public(g)
```

```python
# Create the repository; comment this out* if you run this code more than once :D
# * or convert to markdown!
repo = {'name': my_repo_name, 
        'description': 'I made this repo with an API!', 
        'homepage': '', 
        'private': False, 
        'has_issues': True,
        'has_wiki': True, 
        'has_downloads': False }

rep = None

if repo.get('name'):
    rep = g.create_repo(repo.pop('name'), **repo)

if rep:
    print('Repository {0} was successfully created.'.format(rep.name))
```

```python
# Now save the repository into an object, r
r = g.repository(my_username, my_repo_name)

if r:
    print('Repository {0} is saved as "r".'.format(r.name))

print('What can be done with r:')
public(r)
```

```python
# Now we're going to add two files, README.md and file.md
contents = 'API Test\n======\n\nThis is the README for api-test. It was added via api.'
contents2 = 'This is the text of a file. It was added via api.'

create = r.create_file('README.md', 'this was added via api', contents)
create2 = r.create_file('file.md', 'this was added via api', contents2)

all_the_files = r.contents('/')

print('You have the following files in your repository: ',)
print(all_the_files)
```
